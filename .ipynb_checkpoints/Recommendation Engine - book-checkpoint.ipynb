### Data preparation

```python
del book_data['Unnamed: 0']
```

```python
book_data.head()
```

```python
user_recommendation=book_data.pivot_table(columns='Book.Title',index='User.ID',values='Book.Rating').reset_index()
user_recommendation.head()
```

```python
user_recommendation.index=book_data['User.ID'].unique()
user_recommendation
```

```python
user_recommendation.fillna(inplace=True,value=0,axis=0)
# filling Nan value with 0
```

```python
user_recommendation.dtypes
```

### Model Building and Identifying the similar users for recommedation

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import correlation,cosine
```

```python
user_recommendation.values
```

```python
user_similarity=1-pairwise_distances(user_recommendation.values,metric='cosine')
user_similarity
```

```python
user_df=pd.DataFrame(user_similarity)
user_df
```

```python
user_df.index=book_data['User.ID'].unique()
user_df
```

```python
user_df.columns=book_data['User.ID'].unique()
user_df
```

```python
import numpy as np
np.fill_diagonal(a = user_df.values, val = 0)
# filling diagonal values with 0,since comparing one with same doen't sounds better
```
